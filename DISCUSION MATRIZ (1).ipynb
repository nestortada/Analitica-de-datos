# Matrices de confusión y metricas de rendimiento

```python
from sklearn import datasets
import numpy as np

iris = datasets.load_iris()
X = iris.data[:, [2, 3]]
y = iris.target

print('Class labels:', np.unique(y))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=100.0, solver='lbfgs', multi_class='ovr')
lr.fit(X_train_std, y_train)
```

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, precision_score, recall_score, roc_auc_score, matthews_corrcoef,f1_score
import pandas as pd
import numpy as np
from sklearn. model_selection import train_test_split
from sklearn. linear_model import LogisticRegression
from sklearn import metrics
import matplotlib. pyplot as plt
# Escala las características (importante para la regularización)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

# Valores de C y penalty a considerar
C_values = [0.001, 0.01, 0.1, 1, 10, 100]
penalties = ['l1', 'l2', 'none']



# Itera sobre diferentes combinaciones de C y penalty
for penalty in penalties:
    for C in C_values:
        model = LogisticRegression(penalty=penalty, C=C, solver='saga', random_state=42)
        model.fit(X_train, y_train)
        accuracy = model.score(X_test, y_test)

        print(f'Modelo con penalty={penalty}, C={C}, Precisión={accuracy:.2f}')
        if accuracy >= 1:

          y_pred = lr.predict(X_test_std)
          conf_mat = confusion_matrix(y_true=y_test, y_pred=y_pred)

          sns.heatmap(conf_mat, annot=True, fmt="d", cmap="Blues",
                      xticklabels=['Setosa', 'Versicolor', 'Virginica'],
                      yticklabels=['Setosa', 'Versicolor', 'Virginica'])
          plt.xlabel('Predecidos')
          plt.ylabel('Verdaderos')
          plt.title('Confusion Matrix')
          plt.show()
          recall = recall_score(y_test, y_pred, average='weighted')
          fpr = conf_mat[0, 1] / (conf_mat[0, 0] + conf_mat[0, 1])
          mcc = matthews_corrcoef(y_test, y_pred)
          f1 = f1_score(y_test, y_pred=y_pred, average='weighted')

          print(f"Recuperación (TPR): {recall:.4f}")
          print(f"Tasa de Falsos Positivos (FPR): {fpr:.4f}")
          print(f"Coeficiente de Correlación de Matthews (MCC): {mcc:.4f}")
          print(f"Puntaje F1: {f1:.4f}")

        print("-----"*100)
```



*   A partir de la matriz de confusión, podemos concluir que todos los modelos que tienen una precisión de 1 (exactitud) presentan los mismos valores de FPR, TPR, PRE, F1 y MCC. Esto indica que, independientemente del valor de la penalización o de C, estos modelos poseen la misma capacidad predictiva. Es decir, su rendimiento en términos de predicción es idéntico.



# SVM

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

# To check recent matplotlib compatibility
import matplotlib
from distutils.version import LooseVersion


def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map
    markers = ('o', 's', '^', 'v', '<')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    lab = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    lab = lab.reshape(xx1.shape)
    plt.contourf(xx1, xx2, lab, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # plot class examples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0],
                    y=X[y == cl, 1],
                    alpha=0.8,
                    c=colors[idx],
                    marker=markers[idx],
                    label=f'Class {cl}',
                    edgecolor='black')

    # highlight test examples
    if test_idx:
        # plot all examples
        X_test, y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c='none',
                    edgecolor='black',
                    alpha=1.0,
                    linewidth=1,
                    marker='o',
                    s=100,
                    label='Test set')
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='rbf', random_state=1, gamma=0.2, C=1.0)
svm.fit(X_train_std, y_train)

plot_decision_regions(X_combined_std, y_combined,
                      classifier=svm, test_idx=range(105, 150))
plt.xlabel('Petal length [standardized]')
plt.ylabel('Petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
#plt.savefig('figures/03_15.png', dpi=300)
plt.show()
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='rbf', random_state=1, gamma=100, C=1.0)
svm.fit(X_train_std, y_train)

plot_decision_regions(X_combined_std, y_combined,
                      classifier=svm, test_idx=range(105, 150))
plt.xlabel('Petal length [standardized]')
plt.ylabel('Petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
#plt.savefig('figures/03_15.png', dpi=300)
plt.show()
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='rbf', random_state=1, gamma=500, C=1.0)
svm.fit(X_train_std, y_train)

plot_decision_regions(X_combined_std, y_combined,
                      classifier=svm, test_idx=range(105, 150))
plt.xlabel('Petal length [standardized]')
plt.ylabel('Petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
#plt.savefig('figures/03_15.png', dpi=300)
plt.show()
```



*   A medida que el valor de Gamma aumenta, la precisión del modelo tiende a mejorar. Sin embargo, un Gamma muy alto puede llevar a un sobreajuste, como se observa en las últimas gráficas. Por otro lado, un Gamma bajo puede resultar en un subajuste. Por lo tanto, podemos concluir que la elección adecuada de Gamma es crucial para obtener un buen ajuste del modelo para la predicción. Es importante equilibrar este parámetro para evitar el sobreajuste o el subajuste y así maximizar la capacidad predictiva del modelo.
