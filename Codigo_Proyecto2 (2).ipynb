```python
from scipy.stats import shapiro

variables_cuantitativas = df[['Edad', 'Nivel_Azúcar', 'Presión_Arterial', 'Colesterol', 'Frecuencia_Cardiaca', 'Peso', 'Altura']]
variables_cuantitativas = variables_cuantitativas.astype('float64')

for columna in variables_cuantitativas.columns:
    stat, p_value = shapiro(variables_cuantitativas[columna])
    print(f'Variable: {columna}')
    print(f'Estadístico de prueba: {stat}')
    print(f'Valor p: {p_value}')
    alpha = 0.05

    if p_value > alpha:
        print('No podemos rechazar la hipótesis nula. Los datos parecen seguir una distribución normal.')
    else:
        print('Rechazamos la hipótesis nula. Los datos no siguen una distribución normal.')
        print('-' * 100)
```

```python
import pandas as pd
import scipy.stats as stats

grupo_hombres = df[df['Género'] == 0]['Nivel_Azúcar']
grupo_mujeres = df[df['Género'] == 1]['Nivel_Azúcar']


t_stat, p_val = stats.ttest_ind(grupo_mujeres, grupo_hombres)

alfa = 0.05

if p_val < alfa:
    print("Rechazar la hipótesis nula que las mujeres tienen la misma media de nivel de azucar que los hombres.")
else:
    print(" No hay suficiente evidencia para afirmar que las mujeres tienen el mismo nivel de azucar que los hombres.")

import pandas as pd
import numpy as np
import scipy.stats as stats

grupo_hombres = df[df['Género'] == 0]['Nivel_Azúcar']
grupo_mujeres = df[df['Género'] == 1]['Nivel_Azúcar']


diferencia_medias = grupo_mujeres.mean() - grupo_hombres.mean()
desviacion_std_mujeres = grupo_mujeres.std()
desviacion_std_hombres = grupo_hombres.std()

# Calcular el tamaño de las muestras
n_mujeres = len(grupo_mujeres)
n_hombres = len(grupo_hombres)

error_estandar = np.sqrt((desviacion_std_mujeres**2 / n_mujeres) + (desviacion_std_hombres**2 / n_hombres))

nivel_confianza = 0.95

valor_critico = stats.norm.ppf(1 - (1 - nivel_confianza) / 2)
margen_error = valor_critico * error_estandar

intervalo_confianza = (diferencia_medias - margen_error, diferencia_medias + margen_error)

print(f"Intervalo de confianza para la diferencia de medias: {intervalo_confianza}")
```
