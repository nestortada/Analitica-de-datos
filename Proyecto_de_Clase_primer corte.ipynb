**Pruebas de Normalidad para cada una de las Variables Cuantitativas**

```python
from scipy.stats import shapiro

variables_cuantitativas = mydata.select_dtypes(include=['int64'])

# Realizar la prueba de normalidad de Shapiro-Wilk para cada columna
for columna in variables_cuantitativas.columns:
    stat, p_value = shapiro(variables_cuantitativas[columna])

    print(f'Variable: {columna}')
    print(f'Estadístico de prueba: {stat}')
    print(f'Valor p: {p_value}')

    # Interpretar los resultados
    alpha = 0.05
    if p_value > alpha:
        print('No podemos rechazar la hipótesis nula. Los datos parecen seguir una distribución normal.')
    else:
        print('Rechazamos la hipótesis nula. Los datos no siguen una distribución normal.')
    print('-' * 100)
```

Algunas posibles razones por las que estas variables no son normales son:

**Sesgo:** Algunas variables tienen una distribución sesgada, es decir, que tienen una cola más larga hacia un lado que hacia el otro. Por ejemplo, la variable Income tiene un sesgo positivo, lo que indica que hay más valores altos que bajos.

**Valores atípicos:** Algunas variables tienen valores atípicos, es decir, que se alejan mucho de la media o la mediana. Por ejemplo, la variable Miles tiene algunos valores atípicos en el extremo superior, lo que indica que hay algunos clientes que corren mucho más que el promedio.

**Tamaño de la muestra:** Algunas variables tienen un tamaño de muestra pequeño, es decir, que tienen pocos datos disponibles para analizar. En este caso, la base de datos "CardioGood Fitness" tiene solo 180 observaciones por variable, lo que puede no ser suficiente para capturar la variabilidad de los datos.

Aunque se intenten transformar estas variables a una distribución normal, mediante métodos como Boxcox y Transformaciones Logarítmicas, no es posible convertir estas variables en unas de distribución normal. Por ende, no se pueden hacer pruebas de hipótesis paramétricas para validar supuestos.

# **Matriz de Coeficientes de Variación expresados en porcentajes por cada una de las Variables Cuantitativas**

```python
def coef_vari(datos):

    desviacion_estandar = np.std(datos)
    media = np.mean(datos)

    coef_var = (desviacion_estandar / media) * 100
    return coef_var

piv = pd.pivot_table(mydata, index=['Product'], columns= [], aggfunc=(coef_vari))
print(piv)
print("-"*100)
print("Promedio de Coeficiente de Variación por Producto\n")
print(piv.mean(axis=1))
```

**Variables con mayor variabilidad:** Las variables que tienen un mayor coeficiente de variación son Miles, Usage y Fitness, lo que indica que hay una mayor variación en el número de millas que los clientes esperan correr cada semana, el número de veces que los clientes esperan usar la máquina cada semana y el nivel de estado físico de los clientes.

**Variables con menor variabilidad:** Las variables que tienen un menor coeficiente de variación son Age, Education e Income, lo que indica que hay una menor variación en la edad, el nivel de educación y el ingreso anual de los clientes.

**Productos con mayor variabilidad:** El producto que tiene un mayor coeficiente de variación en general es el TM195, lo que indica que este producto atrae a clientes con características más diversas y heterogéneas.Este producto puede ser el más barato y el más básico de los tres.

**Productos con menor variabilidad:** El producto que tiene un menor coeficiente de variación en general es el TM798, lo que indica que este producto atrae a clientes con características más similares y homogéneas. Este producto puede ser el más caro y el más avanzado de los tres.
